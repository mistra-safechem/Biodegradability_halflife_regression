```python
import pandas as pd
from pandas import read_csv
from pathlib import Path

pd.set_option("display.max_columns", None)
pd.set_option("display.width", 200)
```

```python
working_dir = Path.cwd().parent
raw_data_dir = working_dir / "raw_data"
output_dir = working_dir / "processed_data"

air_data = read_csv(output_dir / "hsbd_air_t_half_data_incl_missing.csv", header=0, sep="\t")
soil_data = read_csv(output_dir / "hsbd_soil_t_half_data_incl_missing.csv", header=0, sep="\t")
water_data = read_csv(output_dir / "hsbd_water_t_half_data_incl_missing.csv", header=0, sep="\t")
sediment_data = read_csv(output_dir / "hsbd_sediment_t_half_data_incl_missing.csv", header=0, sep="\t")

print(f"Before dropping duplicates based on Smiles:")
print(f"Air data: {len(air_data)}")
print(f"Soil data: {len(soil_data)}")
print(f"Water data: {len(water_data)}")
print(f"Sediment data: {len(sediment_data)}")
```

```python
def extract_duplicates(df, subset=["Canonical_smiles"]):
    # find duplicated rows and keep only these, sort by subset columns
    duplicates = df[df.duplicated(subset=subset, keep=False)].sort_values(by=subset)
    return duplicates.sort_values(by=subset)
```

```python
print("duplicates based on Smiles:")
duplicate_air_smiles = extract_duplicates(air_data)
duplicate_soil_smiles = extract_duplicates(soil_data)
duplicate_water_smiles = extract_duplicates(water_data)
duplicate_sediment_smiles = extract_duplicates(sediment_data)
print(f"Air duplicates: {len(duplicate_air_smiles)}")
print(f"Soil duplicates: {len(duplicate_soil_smiles)}")
print(f"Water duplicates: {len(duplicate_water_smiles)}")
print(f"Sediment duplicates: {len(duplicate_sediment_smiles)}")
print("\nduplicates based on Compound name:")
duplicate_air_compound = extract_duplicates(air_data, subset=["Compound"])
duplicate_soil_compound = extract_duplicates(soil_data, subset=["Compound"])
duplicate_water_compound = extract_duplicates(water_data, subset=["Compound"])
duplicate_sediment_compound = extract_duplicates(sediment_data, subset=["Compound"])
print(f"Air duplicates: {len(duplicate_air_compound)}")
print(f"Soil duplicates: {len(duplicate_soil_compound)}")
print(f"Water duplicates: {len(duplicate_water_compound)}")
print(f"Sediment duplicates: {len(duplicate_sediment_compound)}")
```

```python
duplicate_air_smiles.to_csv(output_dir / "dupl_air_data_tmp_smiles_sorted.csv", index=False)
duplicate_soil_smiles.to_csv(output_dir / "dupl_soil_data_tmp_smiles_sorted.csv", index=False)
duplicate_water_smiles.to_csv(output_dir / "dupl_water_data_tmp_smiles_sorted.csv", index=False)
duplicate_sediment_smiles.to_csv(output_dir / "dupl_sediment_data_tmp_smiles_sorted.csv", index=False)

duplicate_air_compound.to_csv(output_dir / "dupl_air_data_tmp_compound_sorted.csv", index=False)
duplicate_soil_compound.to_csv(output_dir / "dupl_soil_data_tmp_compound_sorted.csv", index=False)
duplicate_water_compound.to_csv(output_dir / "dupl_water_data_tmp_compound_sorted.csv", index=False)
duplicate_sediment_compound.to_csv(output_dir / "dupl_sediment_data_tmp_compound_sorted.csv", index=False)
```

```python
print(f"Before dropping duplicates based on Smiles:")
air_len_before = len(air_data)
soil_len_before = len(soil_data)
water_len_before = len(water_data)
sediment_len_before = len(sediment_data)
print(f"Air data: {air_len_before}")
print(f"Soil data: {soil_len_before}")
print(f"Water data: {water_len_before}")
print(f"Sediment data: {sediment_len_before}")

# here a simple drop duplicate should be fine since they should already be dropped.
air_data.drop_duplicates(subset=["Canonical_smiles"], inplace=True)
soil_data.drop_duplicates(subset=["Canonical_smiles"], inplace=True)
water_data.drop_duplicates(subset=["Canonical_smiles"], inplace=True)
sediment_data.drop_duplicates(subset=["Canonical_smiles"], inplace=True)

air_len_after = len(air_data)
soil_len_after = len(soil_data)
water_len_after = len(water_data)
sediment_len_after = len(sediment_data)

print(f"\nAfter dropping duplicates based on Smiles:")
print(f"Air data: {air_len_after} (before: {air_len_before})")
print(f"Soil data: {soil_len_after} (before: {soil_len_before})")
print(f"Water data: {water_len_after} (before: {water_len_before})")
print(f"Sediment data: {sediment_len_after} (before: {sediment_len_before})")
if air_len_after != air_len_before:
    print(f"Warning: Air data has {air_len_before - air_len_after} duplicates based on Smiles.")
if soil_len_after != soil_len_before:
    print(f"Warning: Soil data has {soil_len_before - soil_len_after} duplicates based on Smiles.")
if water_len_after != water_len_before:
    print(f"Warning: Water data has {water_len_before - water_len_after} duplicates based on Smiles.")
if sediment_len_after != sediment_len_before:
    print(f"Warning: Sediment data has {sediment_len_before - sediment_len_after} duplicates based on Smiles.")
```
